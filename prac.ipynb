### Q3-DONE 

```python
#Write a function that uses the uniform distribution (0,1) to generate m samples from a Bernoulli RV with parameter p as input.

m=1000
p=0.7

def bernoulli_Rv():
    graph_data = np.random.uniform(0,1,m)
    bernoulli = np.where(graph_data<p,1,0)
    return bernoulli
bernoulli=bernoulli_Rv()

plt.hist(bernoulli,bins=np.arange(-0.5,2),rwidth=0.9,color='purple',ec='k',linewidth=1)

plt.xticks([0,1])
plt.xlabel('Bernoulli RV')
plt.ylabel('Frequency')
plt.show()
```

## Q-4 DONE

```python
#Write a function (starting from uniform rand (0,1) or preferably the Q3 implementation of the Bern RV) that generates m samples from a geometric RV with parameter p as input

geometric_sams = np.random.randint(m,size=m)

def geo_func():
  for num in range(m):
    k=0
    val=0
    while k==0:
      val+=1
      uniform_Geometric_RV = np.random.uniform(0,1,1)
      if uniform_Geometric_RV[0]<p:
        k=1
    geometric_sams[num] = val
  return geometric_sams

geometric_sams=geo_func()

plt.hist(geometric_sams,bins=np.arange(0.5,geometric_sams.max()+1.5),rwidth=0.7,color='teal',ec='k',linewidth=1.2)

plt.xticks(np.arange(1,geometric_sams.max()+1))

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()   
```

```python
m=1000
alpha=1.5
# alpha=float(input("Enter the value for alpha: "))

def poisson_RV():
    rand_var= np.random.poisson(alpha,m)
    return rand_var

rand_var=poisson_RV()

plt.hist(rand_var,bins=np.arange(-0.5,rand_var.max()+1.5),rwidth=0.8,color='darkred',ec='k',alpha=0.9)

plt.xticks(np.arange(0,rand_var.max()+1))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
