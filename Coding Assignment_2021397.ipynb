**Question-1:** Generate 100 samples that are uniformly distributed over (0,1). Plot a histogram of the samples. Also, plot an empirical CDF.

```python
graph_data = np.random.uniform(0,1,100)

count,bins_count,exceed = plt.hist(graph_data, bins=10,color='lightgreen',ec='k',alpha=0.7,label='Histogram',linewidth=1.2)

bins=bins_count[1:]
pdf = count / sum(count)
# cdf = np.cumsum(pdf)
plt.plot(bins,np.cumsum(pdf),label="CDF",color='r')
plt.legend()
plt.show()
```

**Question-2:** Generate 100 samples that are uniformly distributed over (75, 110). Plot a histogram of the samples. Also, plot an empirical CDF.

**Question-3:** Write a function that uses the uniform distribution (0,1) to generate m samples from a Bernoulli RV with parameter p as input.

```python
def bernoulli_generate():
    graph_data = np.random.uniform(0,1,m)
    bernoulli_func = np.where(graph_data<p,1,0)
    return bernoulli_func

bernoulli = bernoulli_generate()

plt.hist(bernoulli,bins=np.arange(-0.5,2),rwidth=0.9,color='#FD5E53',ec='k',alpha=0.7,linewidth=1)

plt.xticks([0,1])
plt.xlabel('Bernoulli RV')
plt.ylabel('Frequency')
plt.show()
```

**Question-4:** Write a function (starting from uniform rand (0,1) or preferably the Q3 implementation of the Bern RV) that generates m samples from a geometric RV with parameter p as input.

```python
geometric_sams = np.random.randint(m,size=m)

def geo_func():
  for num in range(m):
    k=0
    val=0
    while k==0:
      val+=1
      uniform_Geometric_RV = np.random.uniform(0,1,1)
      if uniform_Geometric_RV[0]<p:
        k=1
    geometric_sams[num] = val
  return geometric_sams

geometric_sams=geo_func()

plt.hist(geometric_sams,bins=np.arange(0.5,geometric_sams.max()+1.5),rwidth=0.7,color='#43D8C9',ec='k',linewidth=1)

plt.xticks(np.arange(1,geometric_sams.max()+1))

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()  
```

**Question-5:** Write a function that generates m samples from a binomial RV with parameters (n,p) as inputs. Starting from uniform rand (0,1) or preferably the Q3 implementation of the Bern RV.

```python
n=int(input("Enter the value for parameter n: ")) #10

binomialf=np.random.randint(m,size=m)

def bern_rv():
    for i in range(m):
        graph_data = np.random.uniform(0,1,n)
        bernoulli_gr = np.where(graph_data<p,1,0)
        binomialf[i] = sum(bernoulli_gr)
    return binomialf

binomialf=bern_rv()

plt.hist(binomialf,bins = np.arange(binomialf.min()-0.5,binomialf.max()+1),rwidth=0.8,color='orange',ec='k',alpha=0.9)
plt.xticks(np.arange(binomialf.min(),binomialf.max()+1))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

**Question-6:** Write a function that generates m samples from a Poisson RV with parameter \alpha as input.

```python
alpha=float(input("Enter the value for alpha: "))

def poisson_RV():
    rand_var= np.random.poisson(alpha,m)
    return rand_var

rand_var=poisson_RV()

plt.hist(rand_var,bins=np.arange(-0.5,rand_var.max()+1.5),rwidth=0.8,color='#EA168E',ec='k',alpha=0.7)

plt.xticks(np.arange(0,rand_var.max()+1))
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

**Question-7:** Write a function that generates m samples from a Gaussian RV with mean and std deviation as input.

```python
mean=float(input("Enter the mean: ")) #2.0
std=float(input("Enter the standard deviation: ")) #0

def generate_samples(stdf,meanf,m):
    return np.random.normal(stdf,meanf,m)

plt.hist(generate_samples(std,mean,m),bins=100,color='darkblue',alpha=0.85)

plt.show()
```

**Question-8:** Write a function that generates m samples from an exponential RV with the rate parameter as input.

```python
def generate_samples(meanf,m):
    return np.random.exponential(meanf,m)

plt.hist(generate_samples(1.0/mean,m),bins=100,color='teal',label='Exponential')
plt.legend()
plt.show()
```

**Question-9:** Generate m=1, 2,..., 1000 samples from the Bernoulli RV (use the function created for the question above) and plot the sample mean for each m.
