```python
#@title  Write a function that generates m samples from a Gaussian RV with mean and std deviation as input. { display-mode: "form" }
from matplotlib import pyplot as mp
import random
import numpy as np

# m=int(input())
# mu=float(input())
# sig=float(input())
m=1000
mu=0.0
sig=2.0
nums=[]
for i in range(m): 
    nums.append(random.gauss(mu, sig)) 
#print(nums)       
mp.hist(nums, bins= 100) 
mp.show()
```

```python
#@title Write a function that generates m samples from an exponential RV with the rate parameter as input. { display-mode: "form" }
from matplotlib import pyplot as mp
from numpy import random 

# m=int(input())
# lda=float(input())
m=1000
lda=2.0

# print(data)
data = random.exponential(1.0/lda, size=m)
mp.hist(data, bins = 100) 
mp.show()
```
